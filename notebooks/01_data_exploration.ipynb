# Data Exploration Notebook

Its purpose is to conduct an exploratory data analysis (EDA) to understand the data's characteristics, identify potential issues, and discover underlying patterns.

## Key Activities:

1. Load and Inspect raw data

2. Check data Quality

3. Visualize the Price Series

4. Analyze Returns

5. Process data for next steps

## 1. Data Loading

```python
import sys

sys.path.append('..')  # Add parent directory to Python path
from src.config import portfolio, data_paths, model_params, results_path

import matplotlib.pyplot as plt

from src.data_processing import load_raw_data, clean_data, add_technical_indicators, save_data

import pandas as pd  # Added for the suggested change

tickers = portfolio["tickers"]

start_date = portfolio["start_date"]

end_date = portfolio["end_date"]

raw_data_path = data_paths["raw_data"]

processed_data_path = data_paths["processed_data"]

for tick in tickers:
    raw_data = load_raw_data(tick, start_date, end_date)
    cleaned_data = clean_data(raw_data)
    # Ensure 'Close' column exists and is a Series
    if 'Close' in cleaned_data.columns and cleaned_data['Close'].ndim == 1:
        cleaned_data['Close'] = pd.to_numeric(cleaned_data['Close'], errors='coerce')
        data_with_indicators = add_technical_indicators(cleaned_data, short=25, long=50)
        save_data(data_with_indicators, f"{processed_data_path}/{tick}_processed.csv")
    else:
        print(f"Warning: 'Close' column missing or malformed for {tick}")
```

## 2. Check Data Quality
## Find outliers, missing values, duplicates

```python
for tick in tickers:

    data = load_raw_data(tick, start_date, end_date)

    print(f"--- {tick} ---")

    print("Missing values:\n", data.isnull().sum())

    print("Duplicates:", data.duplicated().sum())

    print(data.describe())
```

## 3. Visualize the Price Series
## Visualize the closing prices

```python
for tick in tickers:

    data = load_raw_data(tick, start_date, end_date)

    plt.figure(figsize=(14, 7))

    plt.plot(data['Close'], label='Close Price')

    plt.title(f'{tick} Close Price Over Time')

    plt.xlabel('Date')

    plt.ylabel('Close Price')

    plt.legend()

    plt.grid()

    plt.show()
```

## 4. Analyze Returns
## Find returns and plot histograms

```python
for tick in tickers:

    data = load_raw_data(tick, start_date, end_date)

    data['Returns'] = data['Close'].pct_change()

    plt.figure(figsize=(10, 5))

    plt.hist(data['Returns'].dropna(), bins=50, alpha=0.75)

    plt.title(f'{tick} Daily Returns Distribution')

    plt.xlabel('Daily Returns')

    plt.ylabel('Frequency')

    plt.grid()

    plt.show()
```

## 5. Process data for next steps

```python
import pandas as pd
import os

# Combine all processed data into a single DataFrame
combined_data = pd.DataFrame()

for tick in tickers:
    file_path = f"{processed_data_path}/{tick}_processed.csv"
    if os.path.exists(file_path):
        data = pd.read_csv(file_path, index_col=0, parse_dates=True)
        data['Ticker'] = tick
        combined_data = pd.concat([combined_data, data])

# Reset index and sort by date
combined_data = combined_data.reset_index().rename(columns={'index': 'Date'})
combined_data = combined_data.sort_values(['Date', 'Ticker'])

# Save combined data
combined_file_path = f"{processed_data_path}/combined_processed_data.csv"
combined_data.to_csv(combined_file_path, index=False)
print(f"Combined data saved to {combined_file_path}")
print(combined_data.head())
```
